```
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
tf.__version__
```

```
#데이터불러오기
df_mnist_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/mnist_train.csv', header=None)
df_mnist_test  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/mnist_test.csv',  header=None)

df_mnist_train.head()
```

```
#학습데이터 준비
mnist_train_data  = df_mnist_train.iloc[:10000, 1:] / 255
mnist_train_label = df_mnist_train.iloc[:10000, 0]

mnist_test_data  = df_mnist_test.iloc[:1000, 1:] / 255
mnist_test_label = df_mnist_test.iloc[:1000, 0]
```

```
#데이터 라벨 원-핫 인코딩 변환
mnist_train_label_cate = to_categorical(mnist_train_label)
mnist_test_label_cate = to_categorical(mnist_test_label)
mnist_train_label_cate[0]
```

```
#학습모델 설계
model = Sequential()
model.add(Dense(units=784, input_dim=784, activation='sigmoid'))
model.add(Dense(units=10, activation='sigmoid'))
```

```
#학습모델 설정
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['acc'])
```

```
#학습하기
model.fit(mnist_train_data, mnist_train_label_cate, epochs=10)
```

```
#검증
model.evaluate(mnist_test_data, mnist_test_label_cate)
```
